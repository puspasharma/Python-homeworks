# **Projectile Motion**
# Horizontal range;
_ $R = \frac{u^2 sin2\theta}{g}$


# Time of Flight;
 _ $T = \frac{2u sin\theta}{g}$


# Maximum height;
 _ $H = \frac{u^2 sin^2\theta}{2g}$





##Task__1 : Make a class to calculate the range, time of flight, and horizontal range of the projectile fired from the ground.

```
import math
```

```
class Projectile():
    def __init__(self,u,thita,g):
        self.u = u
        self.thita = thita
        self.g = g
    def Horizontal_range(self):
        return (self.u**2) * math.sin(2*self.thita * math.pi/180) / (self.g)
    def Time_of_Flight(self):
        return (self.u*2) * math.sin(self.thita * math.pi/180) / (self.g)
    def Maximum_height(self):
        return (self.u * math.sin(self.thita * math.pi/180))**2 / (self.g*2)
    def updatethita(self,thita):
        self.thita=thita
      
  
```

## Task__2 : Use the list to find the range, time of flight and horizontal range for varying value of angle from 1 degree to 90 degree.

```
import numpy as np
```

```
#initial velocity(u) = 40m/s and we have g=9.8 m/s^2
P = Projectile(40, 0 , 9.8)
R = [] #empty list to collect horizontal range
T = [] #empty list to collect time of flight
H = [] #empty list to collect maximum height
A = [] #empty list to collect angle of projection
for r in range(10,100):
    A.append(r)
    P.updatethita(r)
    r=P.Horizontal_range()
    t=P.Time_of_Flight()
    h=P.Maximum_height()
    R.append(r)
    T.append(t)
    H.append(h)
```

## Task__3 : Make a plot to show the variation of range, time of flight and horizontal range with angle of projection.

```
import matplotlib.pyplot as plt
%matplotlib inline 
plt.subplot(2,2,1)
plt.figure(figsize = [8,8])
plt.title ('Horizontal range vs Angle of Projection')
plt.plot(A,R, label ='Horizontal range')
plt.xlabel('Horizontal range')
plt.ylabel('Angle of projection')
plt.legend()
plt.subplot(2,2,2)
plt.title('Time of Flight vs Angle of Projection',fontsize = 20)
plt.plot(A,T, label = 'Time of Flight')
plt.xlabel('Time of Flight')
plt.ylabel('Angle of Projection')
plt.legend()
plt.subplot(2,2,3)
plt.title('Maximum height vs Angle of Projection',fontsize = 20)
plt.plot(A,H, label = 'Maximum height', color = 'r')
plt.xlabel('Maximum height')
plt.ylabel('Angle of Projection')
plt.legend()
```

```
plt.title('Horizontal range vs Angle of projection')
plt.plot(A,R, label = 'Horizontal range')
plt.xlabel('Angle of projection')
plt.ylabel('Horizontal range')
```
